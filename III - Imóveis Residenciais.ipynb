## Imóveis Residenciais

Importando a biblioteca e os dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head(10)
```

Criando uma lista para agregar os tipos de imóveis sem repetição:

```python
list(dados['Tipo'].drop_duplicates())
```

Selecionando apenas os imóveis que classificamos como residenciais.





```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

Ao passarmos a listagem residencial que desejamos, será verificado dentro do DataFrame se os elementos da nossa lista existem de fato, assinalando True ou False, com "isin()".

```python
dados['Tipo'].isin(residencial).head(10)
```

Atribuindo a nossa lista de imóveis residenciais a variável "selecao".

```python
selecao = dados['Tipo'].isin(residencial)
```

Usaremos um método de seleção no DataFrame, de maneira que só tenhamos os registros que possuem a marcação **True**. 

Criaremos um novo DataFrame: dados_residencial. Em seguida, acionaremos nossa base de dados selecionada.


Como resultado, teremos apenas imóveis do tipo residencial:

```python
dados_residencial = dados[selecao]
dados_residencial
```

Verificando a integridade dos dados em nosso dataframe novo, mais especificamente da variável "Tipo" utilizando o método "drop_duplicates()":

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

Checando o tamanho do DataFrame com o "shape":

```python
dados_residencial.shape[0]
```

Reconstruindo os índices do DataFrame reduzido:

```python
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

## Exportando a Base de Dados

Temos o método "read_csv()" para importação, e para exportação de dados utilizaremos "to_csv()". 

Escreveremos o nome do nosso banco de dados: "dados_residencial", e evocaremos o método de exportação que receberá a pasta de destino e o nome a ser dado para o arquivo csv gerado. 
No caso, chamaremos o novo arquivo de "aluguel_residencial.csv". 

Precisaremos, ainda, especificar o tipo de separador utilizado no banco de dados, o **;**

Obs.: Marcamos o index como falso pois o dataframe já possui.

```python
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep=';', index=False)
```

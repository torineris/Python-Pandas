Utilizando o método "read_csv()" para ler dados do arquivo CSV

## Informações Gerais sobre a Base de Dados

Exibindo os tipos das variáveis existentes na base de dados com o método "dados.dtypes".

Para tornar as informações mais apresentáveis, é utilizado o "pd.DataFrame" para criar um dataframe e visualizar de forma mais elegante.

```python
pd.DataFrame(dados.dtypes)
```

Para melhorar ainda mais a tabela, atribuimos ela a uma variável, além de adicionarmos nomes para as colunas.

```python
pd.DataFrame(dados.dtypes, columns=['Tipos de Dados'])
```

```python
tipos_de_dados = pd.DataFrame(dados.dtypes, columns=['Tipos de Dados'])
tipos_de_dados.columns.name = 'Variáveis'
tipos_de_dados
```

Para verificar a quantidade de registros e de variáveis, é utilizado o "shape"

```python
dados.shape
```

```python
print('A base de dados apresenta {} registros (imóveis) e {} variáveis'.format(dados.shape[0], dados.shape[1]))
```
